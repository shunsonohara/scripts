# ichimill データジェネレータ
ある緯度・経度の範囲でランダムなNMEAフォーマットのデータをVantiqに送信します。
送信データのフォーマットは以下の通り
```json
{
  "deviceId": "<001-005の範囲でランダム>",
  "nmea_gga": "<NMEAフォーマットのGGAセンテンス>",
  "nmea_rmc": "<NMEAフォーマットのRMCセンテンス>"
}
```


````python
# ライブラリインストール
!pip install requests
````

````python
import requests
import random
import time
import json

url = "https://dev.vantiq.co.jp/api/v1/resources/topics/sonohara.jp/inbound"
accesstoken = "<api_token>"

lat_from = 35.653082
lat_to = 35.657714
long_from = 139.757794
long_to = 139.762485


def decimal_to_nmea(decimal, coordinate_type):
    if coordinate_type not in ('latitude', 'longitude'):
        raise ValueError('Invalid coordinate type. Use "latitude" or "longitude".')

    decimal = abs(decimal)
    degrees = int(decimal)
    minutes = (decimal - degrees) * 60

    nmea_value = degrees * 100 + minutes
    return round(nmea_value, 5)


def generate_nmea_gga():
    lat = round(random.uniform(decimal_to_nmea(lat_from, 'latitude'), decimal_to_nmea(lat_to, 'latitude')),
                7)  # Specified range for latitude
    lon = round(random.uniform(decimal_to_nmea(long_from, 'longitude'), decimal_to_nmea(long_to, 'longitude')),
                7)  # Specified range for longitude
    return f"$GNGGA,021119.00,{lat},N,{lon},E,2,10,1.36,50.867,M,39.179,M,1.0,0157*5E\r\n"


def generate_nmea_rmc():
    lat = round(random.uniform(decimal_to_nmea(lat_from, 'latitude'), decimal_to_nmea(lat_to, 'latitude')),
                7)  # Specified range for latitude
    lon = round(random.uniform(decimal_to_nmea(long_from, 'longitude'), decimal_to_nmea(long_to, 'longitude')),
                7)  # Specified range for longitude
    return f"$GNRMC,021119.00,A,{lat},N,{lon},E,0.376,,270723,7.77,W,D,V*51\r\n"


while True:
    # アクセストークン
    authorization = "Bearer" + " " + accesstoken
    headers = {
        'Content-Type': 'application/json',
        'Authorization': authorization
    }
    for i in range(5):
        data = {
            "deviceId": f"{random.randint(1, 5):03}",
            "nmea_gga": generate_nmea_gga(),
            "nmea_rmc": generate_nmea_rmc()
        }
        response = requests.post(url, headers=headers, data=json.dumps(data))
        print(response.status_code, response.reason)
        time.sleep(1)  # wait for 1 second before next POST request
````
